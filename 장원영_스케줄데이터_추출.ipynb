```python
from selenium import webdriver
from bs4 import BeautifulSoup
import time

# Chrome 설정
options = webdriver.ChromeOptions()
options.add_argument('--headless')
options.add_argument('--no-sandbox')
options.add_argument('--disable-dev-shm-usage')
options.add_argument('--disable-gpu')

# 브라우저 실행
driver = webdriver.Chrome(options=options)
driver.get("https://m.cafe.daum.net/IVEstarship/Y8wj?boardType=Q")

# 페이지 로딩 대기
time.sleep(3)  # 웹페이지 랜더링 시간 확보

# 페이지 소스 파싱을 위한 BeautifulSoup 객체 생성
soup = BeautifulSoup(driver.page_source, 'html.parser')

# 추출 조건 설정
condition_1 = "장원영" # "장원영"이 포함된 경우
condition_2 = ["장원영", "이서", "가을", "안유진", "리즈", "레이"] # 포함되서는 안되는 키워드

# 결과를 저장할 리스트
results_1 = []  # "장원영"이 포함된 경우
results_2 = []  # 포함되서는 안되는 키워드가 모두 없는 경우

elements = soup.find_all("strong", class_="tit_subject")

for element in elements:
    text_content = element.get_text(strip=True)

    # 조건 1: "장원영"이 포함된 경우
    if condition_1 in text_content:
        results_1.append(text_content)

    # 조건 2: 제외할 키워드들이 모두 없는 경우
    elif not any(exclude in text_content for exclude in condition_2):
        results_2.append(text_content)

# 결과 출력
print("개인 스케줄:")
print("\n".join(results_1) if results_1 else "스케줄이 없습니다")

print("\n단체 스케줄:")
print("\n".join(results_2) if results_2 else "스케줄이 없습니다")

# 브라우저 닫기
driver.quit()
```
